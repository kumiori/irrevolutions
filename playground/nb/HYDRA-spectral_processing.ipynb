```python
# Testing Pacman
import postprocess as pp    
# import plots as plots

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, inset_axes

from pathlib import Path 
```

```python

def plot_energy(data, parameters):
    
    figure, ax1 = plt.subplots(1, 1, figsize=(8, 6))
    
    ax1.plot(data.load, data.fracture_energy, lw=7, label = 'damage', marker='o')
    ax1.plot(data.load, data.elastic_energy, lw=7, label = 'elastic')
    _total = data.fracture_energy + data.elastic_energy
    ax1.plot(data.load, _total, lw=5, label = 'total', zorder=.1)
    ax1.set_title(
        'Energy of Evolution Map $t\mapsto E_\\ell(y_t)$')

    return figure, ax1

def plot_energy_spectrum(data, parameters):
    
    figure, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))
    
    ax1.plot(data.load, data.fracture_energy, lw=7, label = 'damage', marker='o')
    ax1.plot(data.load, data.elastic_energy, lw=7, label = 'elastic')
    _total = data.fracture_energy + data.elastic_energy
    ax1.plot(data.load, _total, lw=5, label = 'total', zorder=.1)
    ax1.set_title(
        'Energy of Evolution Map $t\mapsto E_\\ell(y_t)$')
    for ax in [ax1, ax2]:
        ax.spines['top'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.spines['right'].set_visible(False)

    ax2 = pp.plot_operator_spectrum(data, params, axis = ax2)
    
    plt.tight_layout()
    
    return figure, (ax1, ax2)
    # ax2.set
```
